* Hay que crear una clase llamada `Encoder` que se encargará de llevar a cabo el cifrado de César con todos los caracteres del código [ASCII](https://elcodigoascii.com.ar). 
* Necesitaremos el atributo `key` para saber con qué clave encriptar el mensaje (un entero del 1 al 25).
* Necesitaremos los métodos de instancia `encrypt` y `decrypt` que se encargarán de cifrar y descifrar, respectivamente.
* `encrypt` tomará por parámetro el mensaje en formato string y lo encriptará según la clave. Devolverá el mensaje encriptado en formato string
* `decrypt` tomará por parámetro el mensaje cifrado en formato string y lo desencriptará según la clave. Devolverá el mensaje original en formato string
* Los espacios en blanco son lo único que no debe ser modificado por el cifrado

PISTA: 
* Para más información acerca de la técnica usada en el cifrado de César, haz click en el siguiente [link](https://es.wikipedia.org/wiki/Cifrado_César)
* Investiga los métodos `chr()` y `ord()` para trabajar con el código ASCII

```
class Encoder():

  def __init__(self, key):
    self.key = key
  
  def encrypt(self, message):
    encrypted = ""
    for c in message:
      if c == " ":
        encrypted += " "
        continue

      encrypted += chr(ord(c) + self.key)

    return encrypted

  def decrypt(self, message):
    original = ""
    for c in message:
      if c == " ":
        original += " "
        continue

      original += chr(ord(c) - self.key)
    
    return original
```

```
encoder = Encoder(key = 3)
```

```
encoder.encrypt("hola")
```
