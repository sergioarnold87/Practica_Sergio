* Hay que crear una clase llamada `Encoder` que se encargará de llevar a cabo el cifrado de César con todos los caracteres del código [ASCII](https://elcodigoascii.com.ar). 
* Necesitaremos el atributo `key` para saber con qué clave encriptar el mensaje (un entero del 1 al 25).
* Necesitaremos los métodos de instancia `encrypt` y `decrypt` que se encargarán de cifrar y descifrar, respectivamente.
* `encrypt` tomará por parámetro el mensaje en formato string y lo encriptará según la clave. Devolverá el mensaje encriptado en formato string
* `decrypt` tomará por parámetro el mensaje cifrado en formato string y lo desencriptará según la clave. Devolverá el mensaje original en formato string
* Los espacios en blanco son lo único que no debe ser modificado por el cifrado

PISTA: 
* Para más información acerca de la técnica usada en el cifrado de César, haz click en el siguiente [link](https://es.wikipedia.org/wiki/Cifrado_César)
* Investiga los métodos `chr()` y `ord()` para trabajar con el código ASCII

```
class Encoder():

  def __init__(self, key):
    self.key = key
  
  def encrypt(self, message):
    encrypted = ""
    for c in message:
      if c == " ":
        encrypted += " "
        continue

      encrypted += chr(ord(c) + self.key)

    return encrypted

  def decrypt(self, message):
    original = ""
    for c in message:
      if c == " ":
        original += " "
        continue

      original += chr(ord(c) - self.key)
    
    return original
```

```
encoder = Encoder(key = 3)
```

* el método `move()` que modificará (o no) la posición en la que se encuentra. Si está en `far` y se modifica, entonces pasará a estar en `near` y viceversa. La decisión de modificar o no la posición será aleatoria.

La clase `Weapon` necesita:

* los atributos:
  - `name`: nombre del arma
  - `damage`: daño que inflige el arma
  - `scope`: alcance del arma. Tiene dos opciones: `short` (distancia corta) o `long` (distancia larga)

```
class Pirate():

  boat_life = 200
  
  def __init__(self, name, weapon):
    self.name = name
    self.weapon = weapon

  def attack(self, target):
    if target.position == "far" and self.weapon.scope == "short":
      print("El ataque no alcanza al enemigo. Haber apuntado mejor")
    else:
      print("{} ataca a {}". format(self.name, target.name))
      target.hit(self.weapon.damage)

  @classmethod
  def hit(cls, damage):
    Pirate.boat_life -= damage
    print("Vida del barco: {}".format("=" * (Pirate.boat_life // 5)))  
```

```
from numpy import random
class Enemy():

  def __init__(self, name, health, position, min_damage, max_damage):
    self.name = name
    self.health = health
    if position not in ["near", "far"]:
      raise ValueError("La posición del enemigo solo puede ser near o far")
    self.position = position
    self.min_damage = min_damage
    self.max_damage = max_damage
  
  def attack(self,target):
    attack_force = random.randint(2)
    if attack_force == 0:
      print("¡Oh no! {} ataca con su ataque mas flojo a {}".format(self.name, target.name))
      Pirate.hit(self.min_damage)

    else:
      print("¡Oh no! {} ataca con su ataque mas fuerte a {}".format(self.name, target.name))
      Pirate.hit(self.max_damage)

  
  def hit(self, damage):
    self.health -= damage

  def move(self):
    if self.position == "near":
      self.position = "far"
    else:
      self.position = "near"
    print("{} ha cambiado su posición a {}".format(self.name, self.position))
```

```
class Weapon():

  def __init__(self, name, damage, scope):
    self.name = name
    self.damage = damage
    if scope not in ["short", "long"]:
      raise ValueError("El alcance del arma solo puede ser short o long")
    self.scope = scope
```
