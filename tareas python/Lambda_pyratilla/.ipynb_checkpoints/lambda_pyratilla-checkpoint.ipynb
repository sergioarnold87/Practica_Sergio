```python
hits = {"brazaleteD": 8, "brazaleteI": 8, "cinturon": 14}

from functools import  reduce

total_hits = reduce(lambda x, y: x + y, hits.values())
```

¡Ayuda a Pyratilla a derrotar al Valrock con los 5 cañones!

Para ello,

> "Pyratilla ha atacado al `target`"

* Con los bucles que creas conveniente deberás atacar y eliminar primero al brazalete derecho, luego al izquierdo y finalmente, al cinturón (es decir, por orden alfabético). Para el orden de ataque, utiliza el método que creas conveniente de la sección 11 y el diccionario `hits` creado anteriormente. 

* En cuanto al ataque, tendrás que tener en cuenta que Pyratilla tiene las mismas posibilidades de acertar como de fallar el tiro con cañón (el pobre no es muy hábil con este tipo de arma a distancia). Por tanto, para determinar si acierta o falla, necesitarás usar el módulo `random`. Puedes considerar fallar el número 0 y acertar, el 1. 
* En caso de que falle, tendrás que indicarlo con el mensaje "¡Oh, no! ¡Pyratilla ha fallado!". En caso de que acierte, deberás usar la función `attack()` con el objetivo correspondiente.
* En cuanto un objetivo haya sido eliminado, deberás indicarlo con pantalla con la frase 

> "¡El `target` ha sido eliminado!"

* Finalmente, cuando el enemigo sea derrotado tendrás que devolver la frase "¡¡¡El Valrock ha sido eliminado!!!" junto al número de cañonazos que Pyratilla ha necesitado para derrotar al Valrock.
* Cada 5 disparos, añade un mensaje que diga:

> "¡Un segundo! ¡Hay que recargar los 5 cañones!"

```python
def attack(target):
    print("Pyratilla ha acabado al", target)
    
```

```python
from numpy import random
targets = sorted(hits.keys())

total_shots = 0

for target in targets:
    total_hits = 0
    
    if target == "brazaleteD":
        objetivo = "brazalete derecho"
    if target == "brazaleteI":
        objetivo = "brazalete izquiero"
    if target == "cinturon":
        objetivo = "cinturon"
        
    while total_hits < hits[target]:
        total_shots += 1
        
        if total_shots % 5 == 0:
            print("\n Un segundo!, ¡Hay que recargar los 5 cañones!\n")
            
        if random.randint(2): #Solo entramos en el if si el numero aleatorio es 1
            attack(objetivo)
            total_hits += 1
        else: #Entramos en el else si el numero aleatorio no es 1 (es cero)
            print("Oh No! Pyratilla ha fallado!!")
        
    print("\n El nivel {} ha sido eliminado! \n\n".format(objetivo))
    
print("\n ¡¡¡Valrock a muerto!!!\n")
print("Se ha realizado {} disparos en total".format(total_shots))        
```

¡Juega con Pyratilla a palabras encadenadas! Las reglas son sencillas:

* La primera letra de tu palabra debe coincidir con la última letra de la palabra anterior
* La palabra que digas no debe haberse dicho antes en la partida

Para jugar,

* debes crear la función `palabras_encadenadas()` que te permita hacer una partida de este juego
* crea una función helper llamada `valid_option()` que compare la última letra de la palabra anterior con la primera de la nueva. Si coinciden, devuelve `True`. De lo contrario, devuelve `False`.
* mientras la palabra sea válida y no haya sido jugada (puedes ayudarte de una lista llamada `played_words`, haz que el usuario introduzca una palabra y que ésta sea comprobada como palabra válida
* si no se cumplen las condiciones, finaliza el juego
* haz que `palabras_encadenadas()` devuelva la lista `played_words`

```python
def valid_option(previous_word, next_word):
    if next_word.lower()[0] == previous_word.lower()[-1]:
        return True
    return False
```

```python
def palabras_encadenadas():
  word = str(input())
  in_game = True
  played_words = [word]

  while in_game:
    next_word = str(input())
    if (valid_option(word, next_word) and next_word not in played_words):
        played_words += [next_word]
        word = next_word
        continue
    else:
        print("Game OveR")
        in_game = False
  return played_words
```
